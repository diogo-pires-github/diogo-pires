```python
import matplotlib.pyplot as plt
```

```python
f1_scores = [0.18941441178321838, 0.302795946598053, 0.257374107837677, 0.14173731207847595, 0.1696225255727768]
models = ['Baseline Model', 'ResNet50V2', 'InceptionV3', 'VGG-19', 'VGG-16']
```

```python
sorted_indices = sorted(range(len(f1_scores)), key=lambda k: f1_scores[k], reverse=True)
sorted_models = [models[i] for i in sorted_indices]
sorted_f1_scores = [f1_scores[i] for i in sorted_indices]

plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_models, sorted_f1_scores, color='#bed62f')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, f'{yval:.3f}', ha='center', va='center', color='black', fontweight='bold')

plt.xlabel('Model Names')
plt.ylabel('F1 Scores')
plt.title('Comparison of F1 Scores for Deep Learning Models')
plt.ylim(0, 0.35) 
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
accuracy_scores = [0.20876911282539368, 0.3181818127632141, 0.2843925952911377, 0.16975060105323792, 0.20152856409549713]

sorted_acc_indices = sorted(range(len(accuracy_scores)), key=lambda k: accuracy_scores[k], reverse=True)
sorted_acc_models = [models[i] for i in sorted_acc_indices]
sorted_accuracy_scores = [accuracy_scores[i] for i in sorted_acc_indices]

plt.figure(figsize=(10, 6))
acc_bars = plt.bar(sorted_acc_models, sorted_accuracy_scores, color='#bed62f')

for bar in acc_bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, f'{yval:.3f}', ha='center', va='center', color='black', fontweight='bold')

plt.xlabel('Model Names')
plt.ylabel('Accuracy')
plt.title('Comparison of Accuracy for Deep Learning Models')
plt.ylim(0, 0.35) 
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
